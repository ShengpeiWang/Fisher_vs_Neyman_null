```python
import pandas as pd
import altair as alt
```

```python
data = pd.read_csv("fisher_vs_neyman.csv", index_col = 0)
data
```

```python
data['permutation_p'] = [2*(0.5 - abs(i - 0.5)) for i in data['permutation_r']]
```

```python
data.groupby('null').mean()
```

```python
data['sig_t'] = data['t-test_p'] < 0.05
data['sig_p'] = data['permutation_p'] < 0.05
```

```python
data.groupby('null')[['sig_t', 'sig_p']].sum()/len(data)
```

```python
alt.Chart(data, title = "Observed difference").mark_area(
    opacity=0.3,
    interpolate='step'
).encode(
    alt.X('sample_effect:Q', bin=alt.Bin(maxbins=100)),
    alt.Y('count()', stack=None),
    alt.Color('null:N')
)
```

```python
alt.Chart(data, title = "P-value from t-test").mark_area(
    opacity=0.3,
    interpolate='step'
).encode(
    alt.X('t-test_p:Q', bin=alt.Bin(maxbins=100)),
    alt.Y('count()', stack=None),
    alt.Color('null:N')
)
```

```python
alt.Chart(data, title = "P-value from permutation test").mark_area(
    opacity=0.3,
    interpolate='step'
).encode(
    alt.X('permutation_p:Q', bin=alt.Bin(maxbins=100)),
    alt.Y('count()', stack=None),
    alt.Color('null:N')
)
```

```python
alt.Chart(data, title = "Correlation between p-values from t-test vs permutation")\
.mark_point(
).encode(
    alt.X('permutation_p:Q'),
    alt.Y('t-test_p:Q'),
    alt.Color('null:N')
)
```
